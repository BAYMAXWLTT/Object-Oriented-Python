```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(a,b,x):
    return a*x+b
def gradient_a(a,b,x,y):
    return 2*(a*x+b-y)*x
def gradient_b(a,b,x,y):
    return 2*(a*x+b-y)
```

```python
points=[[i,i] for i in range(30)]
for v,_ in enumerate(points):
    perturbation=random.randrange(0,4)
    if v%2==0:
        points[v][1]+=perturbation
    else:
        points[v][1]-=perturbation
    
x=np.linspace(0,30,100)
args=np.random.rand(1,2)[0]
print(args)

for index,point in enumerate(points):
    a,b=args[0],args[1]
    _x,_y=point[0],point[1]
    g_a=gradient_a(a,b,_x,_y)
    g_b=gradient_b(a,b,_x,_y)
    args[0]=args[0]-0.0002*g_a
    args[1]=args[1]-0.0002*g_b
    y=[f(args[0],args[1],x_i) for x_i in x]
    if index%6==0 or index == len(points):
        plt.plot(x,y,label='y = %.3f x + %.3f'%(args[0],args[1]))

plt.legend()
plt.scatter([point[0] for point in points],[point[1] for point in points])
```
